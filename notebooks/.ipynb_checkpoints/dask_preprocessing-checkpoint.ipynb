```python
import sys
import pandas as pd
import pyarrow
import pyarrow.parquet
from tqdm import tqdm
import gc

sample = pd.read_csv('../datasets/taxi_1B.csv', nrows=1000000)
dtypes = {col: sample[col].dtype for col in sample.columns}
SPLIT_ROWS = 1_000_000
```

```python
!mkdir -p ../datasets/taxi_parquet
```

```python

for i,chunk in enumerate(tqdm(pd.read_csv('../datasets/taxi_1B.csv', chunksize=SPLIT_ROWS))):
    for col in chunk.columns:
        if chunk[col].dtype != dtypes[col]:
            chunk[col] = chunk[col].astype(dtypes[col])
    chunk.to_parquet(f"../datasets/taxi_parquet/data_{i}.parquet", engine='pyarrow')
    del chunk
    gc.collect()
```

```python
from glob import glob
SPLIT_ROWS = 1_000_000
files = sorted(list(glob('../datasets/taxi_parquet/*')),key=lambda x: int(x.split('/')[-1].split('.')[0].split('_')[1]))
skip_rows = SPLIT_ROWS *len(files)
```

```python
i = len(files)
for chunk in tqdm(pd.read_csv('../datasets/taxi_1B.csv', chunksize=SPLIT_ROWS, skiprows=skip_rows, names=sample.columns)):
    for col in chunk.columns:
        if chunk[col].dtype != dtypes[col]:
            chunk[col] = chunk[col].astype(dtypes[col])
    chunk['payment_type'] = chunk['payment_type'].astype(str)
    chunk.to_parquet(f"../datasets/taxi_parquet/data_{i}.parquet", engine='pyarrow')
    i+=1
    del chunk
    gc.collect()
```
