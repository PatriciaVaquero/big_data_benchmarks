```python
index = 'Unnamed: 0'
vaex_results = pd.read_csv("s3://vaex-sagemaker-demo/benchmarks/vaex_1b_c5d2xlarge_results.csv").dropna()
spark_results = spark_results.rename(columns={'filter':'filter data','standard':'standard deviation'}).dropna()
dask_results = pd.read_csv("s3://vaex-sagemaker-demo/benchmarks/dask_1b_c5d2xlarge_results.csv").rename(columns={'standard':'standard deviation'}).dropna()
turicreate_results = pd.read_csv("s3://vaex-sagemaker-demo/benchmarks/turicreate_1b_c5d2xlarge_results.csv").rename(columns={'standard':'standard deviation'}).dropna()
datatable_results = pd.read_csv("s3://vaex-sagemaker-demo/benchmarks/datatable_1b_c5d2xlarge_results.csv").rename(columns={'standard':'standard deviation'}).dropna()

spark_results['Unnamed: 0'] = spark_results['Unnamed: 0'].apply(lambda x: x if x!='standard' else 'standard deviation')
spark_results['Unnamed: 0'] = spark_results['Unnamed: 0'].apply(lambda x: x if x!='filter' else 'filter data')
dask_results['Unnamed: 0'] = dask_results['Unnamed: 0'].apply(lambda x: x if x!='standard' else 'standard deviation')
turicreate_results['Unnamed: 0'] = turicreate_results['Unnamed: 0'].apply(lambda x: x if x!='standard' else 'standard deviation')
datatable_results['Unnamed: 0'] = datatable_results['Unnamed: 0'].apply(lambda x: x if x!='standard' else 'standard deviation')


df = pd.merge(vaex_results, spark_results, on=index, how='outer')
df = pd.merge(df, dask_results, on=index, how='outer')
df = pd.merge(df, turicreate_results, on=index, how='outer')
df = pd.merge(df, datatable_results, on=index, how='outer')
df = df.rename(columns={'Unnamed: 0':'function'})
df
```

```python
df.to_csv('results.csv')
```
